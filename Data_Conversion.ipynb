```python
import os
import pandas as pd
import numpy as np
```

```python
file_path = 'arxiv-metadata-oai-snapshot.json'

print(f'ArXiv Metadata File Size: {os.stat(file_path).st_size / (1024**2)} MB')
records = 0
with open(file_path, 'r', encoding='utf-8') as data_json:
    for record in data_json:
        records += 1
        
print('Number of Records in Data: ', records)
```

```python
#load file into data frame
with open(file_path, 'r', encoding='utf-8') as data_json:
    data = pd.read_json(data_json, lines = True)
```
