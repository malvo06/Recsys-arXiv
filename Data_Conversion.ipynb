### Convert DBLP Data into a Graph

```python
data = data.fillna('0')
```

```python
data_ref = data.T.to_dict(orient='list') 
```

```python
dict_nodes = {v[0]:v[4] for k, v in data_ref.items()}
data_ref=0
```

```python
dict_nodes
```

```python
import networkx as nx
from networkx.readwrite import json_graph
```

```python
G = nx.from_dict_of_lists(dict_nodes)
```

```python
print(nx.info(G))
```

```python
data.to_pickle('dblp_v10_0_titles.pkl')
```
